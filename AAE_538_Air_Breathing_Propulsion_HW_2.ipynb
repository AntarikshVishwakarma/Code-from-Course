```
import matplotlib.pyplot as plt
 # declaring sea level atmosphere condition
p = 101325
gamma = 1.4
 
def stagp(sp,M,gamma): # defining a function for stagnation pressure
  p0 = sp*pow((1+((gamma-1)/2)*M**2),(gamma/(gamma -1)))
  return p0
# declaring one line functions for N and dNdM  
N = lambda i: (i*pow((1+((gamma-1)/2)*i**2),0.5)/(1+gamma*i**2)) 
dNdM = lambda M: N(M)*(1/M + ((gamma-1)/2)*(M/(1+((gamma-1)/2)*M**2)) 
 - (2*gamma*M)/(1+gamma*M**2))

import numpy

Asp =[]
mp = []
Machn = []
 
for i in numpy.arange(0,3.05,0.05):
  Machn.append(i)
  Asp.append(stagp(p,i,gamma)) # applying the conditions here
  if i <= 1: 
  # if M < 1 then actual stagnation pressure = measured stagnation pressure
    mp.append(stagp(p,i,gamma))
  elif i>1 :
    M = []
    M.append(1/i)
    j = 0
    N1 = N(i)   
    error = 1
    while error >= pow(10,-14): # computing M behind the shock
      M.append(M[j] +(N1 - N(M[j]))/dNdM(M[j]))
      k = j+1
      error = abs(M[k] - M[j])
      j+=1
    l=j-1  
    p2 = p*((1+gamma*i**2)/(1+gamma*M[l]**2)) # using jump conditions 
    mp.append(stagp(p2,M[l],gamma))
    
plt.figure(figsize=(10,5))
plt.scatter(Machn, Asp, color='r', marker = 'o',
            label = "Actual Stagnation Pressure")
plt.scatter(Machn, mp, color='b', marker = 'D', 
            label = "Pressure measured by probe")
plt.title('Pressure as a function of Mach # \n',fontsize = 25)
plt.xlabel('M', fontsize = 20)
plt.ylabel('Stagnation Pressure (Pa)', fontsize = 20)
plt.legend(loc="upper left")
plt.grid()
plt.show();
```

```
import numpy
import math

gamma = 1.4

p = 101325
T = 300
A1 = A2 = 0.1
ha = [0,1,2,4,6] # given values 
stagTratio =[]
R = 287

N = lambda i: (i*pow((1+((gamma-1)/2)*i**2),0.5)/(1+gamma*i**2))
dNdM = lambda M: N(M)*(1/M + (((gamma-1)/2)*(M/(1+((gamma-1)/2)*(M**2))))
 - ((2*gamma*M)/(1+gamma*(M**2))))
G = lambda M : (1+gamma*M**2)/(pow((1+((gamma-1)/2)*M**2),gamma/(gamma-1)))

# calulating sqrt(T02/T01)
for i in range(0,len(ha),1):
  stagTratio.append(math.sqrt(1+ha[i]))

# declaring empty matrices
velocity = numpy.empty([180,len(stagTratio)], order='C')
massflow = numpy.empty([180,len(stagTratio)], order='C')
backp = numpy.empty([180,len(stagTratio)], order='C')
Mentry = numpy.empty([180,len(stagTratio)], order='C')
Mexit = numpy.empty([180,len(stagTratio)], order='C')
frontp = numpy.empty([180,len(stagTratio)], order='C')

# looping through sqrt(T02/T01)
for j in range(0,len(stagTratio),1):
  e = 0
  Me = []
  bp = []
  for i in numpy.arange(0.1,1,0.005):
    Mexit[e][j] = i
    error = 1
    M=[]
    M.append(0.1)
    a = 0
    while error >= pow(10,-10):
 # Modifying the Newton's algorithm to calculate for M1     
      M.append(M[a] +((N(i)*(1/(stagTratio[j])) - N(M[a]))/dNdM(M[a])))
      k = a+1
      error = abs(M[k] - M[a])
      a+=1
    l = a-1
    M1 = M[l]
    # isentropic relation for pressure
    p1 = p/(pow(1+((gamma-1)/2)*M1**2,gamma/(gamma-1))) 
    # isentropic relation for temperature
    T1 = T/(1+((gamma-1)/2)*M1**2)
    a1 = math.sqrt(gamma*R*T1)
    v1 = M1*a1 # finding velocity
    velocity[e][j] = v1
    frontp[e][j] = p1
    pnot2 = p*(G(M1)/G(i))
    p2 = pnot2/(pow(1+((gamma-1)/2)*i**2,gamma/(gamma-1)))
    Mentry[e][j] = M1
    Me.append(M1)
    bp.append(p2)
    backp[e][j] = p2
    e+=1
  row = 0
  # finding mass flow for each exit M
  for b in range(0,len(bp),1):
    m = (bp[b]*N(Me[b])*math.sqrt((1+((gamma-1)/2)*M2**2))*A2*
         math.sqrt(gamma/R)*Mexit[row][j])/(N(Mexit[row][j])*math.sqrt(T))
    massflow[row][j]=m
    row+=1


plt.figure(figsize=(10,5))
for d in range(0,len(stagTratio),1):
  plt.plot(backp[:,d], massflow[:,d],label = '$\Delta T$ / T = %s' %ha[d])
  
plt.title('Mass flow as a function of back pressure \n',fontsize = 25)
plt.xlabel('$P_2$(Pa)', fontsize = 20)
plt.ylabel('$\dot m$ (kg/s)', fontsize = 20)
plt.legend(loc="upper left")
plt.grid()
plt.show();

plt.figure(figsize=(10,5))
for d in range(0,len(stagTratio),1):
  plt.plot(backp[:,d], Mentry[:,d],label = '$\Delta T$ / T = %s' %ha[d])
  
plt.title('Mach# at station 1 as a function of back pressure \n',fontsize = 25)
plt.xlabel('$P_2$(Pa)', fontsize = 20)
plt.ylabel('$M_1$', fontsize = 20)
plt.legend(loc="upper left")
plt.grid()
plt.show();

plt.figure(figsize=(10,5))
for d in range(0,len(stagTratio),1):
  plt.plot(backp[:,d], Mexit[:,d],label = '$\Delta T$ / T = %s' %ha[d])
  
plt.title('Mach# at station 2 as a function of back pressure \n',fontsize = 25)
plt.xlabel('$P_2$(Pa)', fontsize = 20)
plt.ylabel('$M_2$', fontsize = 20)
plt.legend(loc="upper right")
plt.grid()
plt.show();

plt.figure(figsize=(10,5))
for d in range(0,len(stagTratio),1):
  plt.plot(backp[:,d], frontp[:,d],label = '$\Delta T$ / T = %s' %ha[d])
  
plt.title('Pressure at station 1 as a function of back pressure \n',
          fontsize = 25)
plt.xlabel('$P_2$(Pa)', fontsize = 20)
plt.ylabel('$P_1$(Pa)', fontsize = 20)
plt.legend(loc="lower right")
plt.grid()
plt.show();

plt.figure(figsize=(10,5))
for d in range(0,len(stagTratio),1):
  plt.plot(backp[:,d], velocity[:,d],label = '$\Delta T$ / T = %s' %ha[d])
  
plt.title('Velocity at station 1 as a function of back pressure \n',
          fontsize = 25)
plt.xlabel('$P_2$(Pa)', fontsize = 20)
plt.ylabel('$v_1$(m/s)', fontsize = 20)
plt.legend(loc="upper right")
plt.grid()
plt.show();

```

As the back pressure is specified (by setting the exit Mach #). Heat addition decreases stagnation pressure. From the isentropic relation between stagnation and static pressure we can say that for a fixed static pressure and ratio of specific heat with decrease in stagnation pressure the Mach # should decrease. 

In problem 3 the back pressure is not specified, hence we do not see a decrease in Mach # with increase in heat addition.
